```python
# find all objects in image (having certain color) and label them
# find largest object
# find bounding box
# expand bounding box
# draw bounding box

import cv2
import numpy as np

img = cv2.imread("sampleAll.JPG", 1)
img = cv2.GaussianBlur(img, (3, 3), 0)

h, s, v = cv2.split(cv2.cvtColor(img, cv2.COLOR_BGR2HSV))

result = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
r = img.copy()

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        #if h[i, j] < 4: # red
        #if h[i, j] > 108 and h[i, j] < 112: # blue
        if h[i, j] > 20 and h[i, j] < 24: # yellow
            result[i, j] = 255
        else:
            result[i, j] = 0
            r[i, j, 0] = 0
            r[i, j, 1] = 0
            r[i, j, 2] = 0
            
kernel = np.ones((3,3),np.uint8)
result = cv2.erode(result,kernel,iterations = 1)
result = cv2.morphologyEx(result, cv2.MORPH_OPEN, kernel)

kernel = np.ones((5,5),np.uint8)
result = cv2.dilate(result,kernel,iterations = 1)

width = img.shape[1]
height = img.shape[0]

labels = np.zeros((height, width, 1), np.uint8)
label = 0

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if result[i,j] == 255 and labels[i, j] == 0:
            label += 1
            Q = []
            labels[i,j] = label
            Q.append((i, j))
            while(Q != []):
                qi, qj = Q.pop(0)
                # i-1 j-1
                if(qi > 0):
                    if(qj > 0):
                        if(result[qi - 1, qj - 1] == 0 and labels[qi - 1, qj - 1] == 0):
                            labels[qi - 1, qj - 1] = label
                            Q.append((qi - 1, qj - 1))
                            
                # i-1 j
                    if(result[qi - 1, qj] == 0 and labels[qi - 1, qj] == 0):
                        labels[qi - 1, qj] = label
                        Q.append((qi - 1, qj))
                # i-1 j+1
                    if(qj < width - 1):
                        if(result[qi - 1, qj + 1] == 0 and labels[qi - 1, qj + 1] == 0):
                            labels[qi - 1, qj + 1] = label
                            Q.append((qi - 1, qj + 1))
                # i j-1
                if(qj > 0):
                    if(result[qi, qj - 1] == 0 and labels[qi, qj - 1] == 0):
                        labels[qi, qj - 1] = label
                        Q.append((qi, qj - 1))
                # i j+1
                if(qj < width - 1):
                    if(result[qi, qj + 1] == 0 and labels[qi, qj + 1] == 0):
                        labels[qi, qj + 1] = label
                        Q.append((qi, qj + 1))
                # i+1 j-1
                if(qi < height - 1):
                    if(qj > 0):
                        if(result[qi + 1, qj - 1] == 0 and labels[qi + 1, qj - 1] == 0):
                            labels[qi + 1, qj - 1] = label
                            Q.append((qi + 1, qj - 1))
                # i+1 j
                    if(result[qi + 1, qj] == 0 and labels[qi + 1, qj] == 0):
                        labels[qi + 1, qj] = label
                        Q.append((qi + 1, qj))
                # i+1 j+1
                    if(qj < width - 1):
                        if(result[qi + 1, qj + 1] == 0 and labels[qi + 1, qj + 1] == 0):
                            labels[qi + 1, qj + 1] = label
                            Q.append((qi + 1, qj + 1))
                        
            

cv2.imshow('Signs', result)
cv2.imshow('Red', r)
cv2.imshow('Labels', labels)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
